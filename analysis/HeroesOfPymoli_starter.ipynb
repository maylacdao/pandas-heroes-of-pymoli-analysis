```python
grouped = pd.DataFrame(renamed_data['Price'].groupby([renamed_data['Gender'], renamed_data['SN']]).sum()).reset_index()
grouped.head()
```

```python
df2 = pd.DataFrame(grouped.groupby(['Gender'])['Price'].mean()).reset_index().round(2)
df2
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_dem = pd.cut(renamed_data['Age'],8, labels=["<10", "10-14", "15-19","20-24","25-29","30-34","35-39","40+"]).value_counts()
age_dem
```

```python
df5 = pd.DataFrame(age_dem).reset_index().rename(columns={"index":"Age Group","Age":"Total Count"}).sort_values(by=['Age Group'], ascending=True)
df6 = df5.set_index('Age Group')
df6
```

```python
df7 = df6.apply(lambda x: 100 * x / float(x.sum())).rename(columns={"Total Count":"Percentage of Players"}).round(2).astype(str) + '%'
df7
```

```python
age_demographics = pd.concat([df6, df7], axis=1)
age_demographics
```
